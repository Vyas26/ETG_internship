### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
dataf = pd.read_csv("summer.csv")
dataf = dataf[dataf['Medal'] == 'Gold']

data=[]

for sport in dataf['Sport'].unique():
  data.append([sport,len(dataf[dataf['Sport']==sport])])

pd.DataFrame(data,columns=['Sport','freq']).sort_values(by='freq',ascending=False).head().plot(x='Sport',y='freq',kind='bar',figsize=(10,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport')['Medal'].count().sort_values(ascending=False).head().plot.bar(figsize=(8,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete')['Medal'].count().sort_values(ascending=False).head().plot.bar(figsize=(5,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
dataf = pd.read_csv("summer.csv")
dataf = dataf[dataf['Medal'] == 'Gold']

data=[]

for athlete in dataf['Athlete'].unique():
  data.append([athlete,len(dataf[dataf['Athlete']==athlete])])

pd.DataFrame(data,columns=['Athlete','freq']).sort_values(by='freq',ascending=False).head().plot(x='Athlete',y='freq',kind='bar',figsize=(10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df = pd.read_csv("summer.csv")
ind=df[df['Country']=='IND']
gold=ind[ind['Medal']=='Gold']

first_medal=pd.DataFrame(gold,columns=['Year']).sort_values(by='Year',ascending=True).head(1)

for i in first_medal['Year']:
  print(i)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df = pd.read_csv("summer.csv")

data =[]

for event in df['Event'].unique():
  data.append([event,len(df[df['Event']==event])])

pd.DataFrame(data,columns=['Event','freq']).sort_values(by='freq',ascending=False).head().plot(x='Event',y='freq',kind='bar',figsize=(10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df = pd.read_csv("summer.csv")
df=df[df['Medal']=='Gold']
df=df[df['Gender']=='Women']

data = []

for sport in df['Sport'].unique():
  data.append([sport,len(df[df['Sport']==sport])])

pd.DataFrame(data,columns=['Sport','freq']).sort_values(by='freq',ascending=False).head().plot(x='Sport',y='freq',kind='bar',figsize=(10,5))
```
